```python
import pandas as pd
data=pd.read_csv('sportdata.csv')
```

```python
data['Sex'].replace(['Female','Male'],[0,1],inplace=True)
data.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
x=data.drop(columns='Sport')
x.head()
y=data.drop(columns=['Age','Height','Sex'])
y.head()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#BUILD ML MODELS
clf=DecisionTreeClassifier()
#TRAINING THE MODELS
clf.fit(x_train,y_train)
```

```python
test=clf.predict(x_test)
test=clf.predict([[13,3,1]])
test
```

```python
import joblib as jb
jb.dump(clf,'sport_model.joblib')
```

```python
from sklearn.tree import DecisionTreeClassifier
import joblib as jb
clf=jb.load('sport_model.joblib')
test=clf.predict([[13,3,1]])
test
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy=accuracy_score(y_test,test)
```
